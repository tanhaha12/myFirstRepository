```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df["Label"] = np.where(main_df["Class"] == 1, "Fraud", "No Fraud")
```

```python
# Plotting feature vs feature, v1 vs v2
sns.scatterplot(x = "V1", y = "V2", data = main_df, alpha = 0.4, hue = "Label")
```

```python
df_features = main_df.iloc[:,1:29]
df_features.head()
```

```python
n_clust = 5
kmeans = KMeans(n_clusters = n_clust)
kmeans.fit(df_features)
labels = kmeans.labels_
```

```python
plt.hist(labels)
```

```python
import numpy as np

for clusterNum in range(n_clust):
    nums = np.where(labels == clusterNum)[0]  # Ensure indices are accessed properly
    cluster_df = main_df.iloc[nums]
    means = cluster_df.mean(numeric_only=True)  # Explicitly set numeric_only to True
    percent_fraud = means["Class"]
    print("Cluster {:2d} contains {:.1%} fraud".format(clusterNum, percent_fraud))
```
